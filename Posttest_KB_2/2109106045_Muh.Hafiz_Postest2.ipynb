```python
import pandas as pd
df = pd.read_csv('CarFuelandEmissions.csv')
```

## Dictionary

```python
dictionary = {}
dictionary["Jumlah_Record"] = df.shape
dictionary["Nama_Atribut"] = list(df.columns)
dictionary["Atribut_Objek"] = list(df.select_dtypes(object))
dictionary["Atribut_Angkah"] = list(df.select_dtypes(['float64']).columns)
dictionary["Atribut_Null"] = list(df.loc[:, df.isnull().any()].columns)
```

## jumlah record

```python
print(f"Jumlah Record: {df.shape}")
```

```python
df.info()
```

## Nama Atribut

```python
nomor = 1
for i in dictionary['Nama_Atribut']:
    print(f"{nomor}.{i}")
    nomor +=1
```

## Jumlah Atribut

```python
nomor = 1
for i in dictionary["Atribut_Objek"]:
    print(f"{nomor}.{i}")
    nomor +=1
jumlah = dictionary["Atribut_Objek"]
print("")
print("Jumlah Atribut: ",len(jumlah))
```

## Jumlah Atribut Tipe Angka

```python
nomor = 1
for i in dictionary["Atribut_Angkah"]:
    print(f"{nomor}.{i}")
    nomor +=1;
jumlah_angka = dictionary["Atribut_Angkah"]
print('')
print("Jumlah AtribUt Bertipe Angka: ", len(jumlah_angka))
```

## Jumlah Atribut Yang Memiliki Nilai Null

```python
no = 1
for i in dictionary["Atribut_Null"]:
    print(f"{no}.{i}")
    no +=1
jumlah_null = dictionary["Atribut_Null"]
print("")
print("Jumlah Atribut Null:", len(jumlah_null) )
                    
```
