```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('car data.csv')
df
```

# GRAFIK BATANG

```python
plt.figure(figsize=(5,18))
x = list(df["Year"])
y = list(df["Car_Name"])

plt.xlabel("customer")
plt.ylabel("penjualan")
plt.title("Grafik Batang")
plt.bar(x,y,color="red")
plt.show()
```

# Grafik Box Plot

```python
plt.figure(figsize=(5,18))
plt.title('Grafik Boxplot')
sns.boxplot(data=df,x="Year",y="Car_Name")
plt.show()
```

# GRAFIK GARIS DAN FLOT

```python
plt.figure(figsize=(25,10))
x = list(df["Year"])
y = list(df["Present_Price"])
plt.title("Grafik Garis")
plt.plot(x,y)
plt.show()
```

# GRAFIK PIE

```python
data_pie = df["Year"].value_counts()
data_pie = df["Year"].value_counts().rename_axis("Year")
data_pie = df["Year"].value_counts().rename_axis("Year").reset_index(name="NO")
data_pie.head(10)
```

```python
plt.figure(figsize=(10,20))
plt.title("Grafik Pie Merek Mobil Pada Tahun Yang Berbeda")
plt.pie(data_pie["NO"], labels=data_pie["Year"], startangle=90, autopct="%.2f%%")
plt.show()
```

# Grafik Scatter Plot

```python
plt.figure(figsize=(10,10))
plt.xlabel("Harga")
plt.ylabel("Tahun")
plt.scatter(x=df["Selling_Price"], y=df["Year"])
plt.show()
```
